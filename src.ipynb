```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
weight = np.random.normal(loc=70, scale=5, size=500)
#height = np.random.normal(loc=170, scale=7.5, size=500)
height = (-2*weight) + 5 + np.random.normal(loc=20, scale=5, size=500)
data=pd.DataFrame(data={'Weight (Kgs)': weight, 'Height (cms)': height})
```

```python
data
```

```python
data.mean(axis=0)
```

```python
plt.scatter(data['Weight (Kgs)'], data['Height (cms)'])
plt.xlabel('Weight (Kgs)')
plt.ylabel('Height (cms)')
plt.title('Scatter plot of Weight vs Height')
plt.show()
```
